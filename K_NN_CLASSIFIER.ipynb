```python
gender=pd.get_dummies(df['Gender'],drop_first=True)
df=pd.concat([df,gender],axis=1)
df.head()
```

```python
df.drop(['Gender'],axis=1,inplace=True)
```

```python
X=df[['Age','EstimatedSalary','Male']]
y=df['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
```

```python
scaled_features=scaler.transform(X)
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features,columns=X.columns)
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scaled_features,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,ConfusionMatrixDisplay
print("CLASSIFICATION REPORT ",classification_report(y_test,pred))
print("ACCURACY:",accuracy_score(y_test,pred))
print("PRECISION:",precision_score(y_test,pred))
print("RECALL:",recall_score(y_test,pred))
```
